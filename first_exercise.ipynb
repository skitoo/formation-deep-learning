```python
from first_exercise import create_dataset, model_training, model
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset, expected = create_dataset()
```

```python
plt.scatter(dataset[:,0], dataset[:,1], c=expected, cmap="summer")
plt.show()
```
